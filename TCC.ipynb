```python
# Análise Preditiva de Resultados Olímpicos de Natação
# Adaptado para Google Colab

# Este notebook contém o código completo para análise preditiva dos resultados olímpicos de natação
# com base nos resultados dos campeonatos mundiais anteriores.

# Autor: [Enzo Novaes]
# Data: Julho 2025

# %% [markdown]
# # Data Science na Natação – Prevendo resultados das Olimpíadas através dos resultados dos Campeonatos Mundiais
#
# Este notebook implementa uma análise preditiva dos resultados olímpicos na natação com base nos resultados dos campeonatos mundiais anteriores. Utilizando técnicas de Data Science e modelos de aprendizado de máquina, o estudo busca estabelecer relações estatísticas entre os desempenhos dos atletas em campeonatos mundiais e suas subsequentes performances nas Olimpíadas.

# %% [markdown]
# ## 1. Configuração do Ambiente

# %%
# Instalação dos pacotes necessários
!pip install pandas numpy scikit-learn xgboost matplotlib seaborn

# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import os
from google.colab import files
import io

# Configuração de visualização
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

# %% [markdown]
# ## 2. Upload e Carregamento dos Dados
#
# Nesta seção, você precisará fazer o upload dos arquivos Excel contendo os dados das competições de natação.

# %%
# Instruções para upload dos arquivos
print("Por favor, faça o upload dos arquivos Excel com os dados das competições de natação.")
print("1. Olympic_Results.xlsx - Resultados olímpicos até Paris 2024")
print("2. Aquatics_Championship_Results.xlsx - Resultados dos campeonatos mundiais de 1973 a 2023")

# Upload dos arquivos
uploaded_olympic = files.upload()  # Aguarde o upload do arquivo Olympic_Results.xlsx
uploaded_championship = files.upload()  # Aguarde o upload do arquivo Aquatics_Championship_Results.xlsx

# Verificação dos arquivos carregados
olympic_file_name = list(uploaded_olympic.keys())[0]
championship_file_name = list(uploaded_championship.keys())[0]

print(f"Arquivo olímpico carregado: {olympic_file_name}")
print(f"Arquivo de campeonatos carregado: {championship_file_name}")

# %% [markdown]
# ## 3. Exploração e Pré-processamento dos Dados

# %%
# Carregamento dos dados
olympic_data = pd.read_excel(io.BytesIO(uploaded_olympic[olympic_file_name]))
championship_data = pd.read_excel(io.BytesIO(uploaded_championship[championship_file_name]))

# Exibição das primeiras linhas de cada conjunto de dados
print("Primeiras linhas dos dados olímpicos:")
display(olympic_data.head())

print("\nPrimeiras linhas dos dados dos campeonatos mundiais:")
display(championship_data.head())

# Informações sobre os conjuntos de dados
print("\nInformações sobre os dados olímpicos:")
print(olympic_data.info())

print("\nInformações sobre os dados dos campeonatos mundiais:")
print(championship_data.info())

# %%
# Função para converter tempos do formato MM:SS.ms para segundos
def convert_Results_to_seconds(Results_str):
    if pd.isna(Results_str):
        return np.nan

    try:
        # Verifica se o tempo já está em segundos
        if isinstance(Results_str, (int, float)):
            return float(Results_str)

        # Converte de string para segundos
        Results_str = str(Results_str).strip()

        # Verifica diferentes formatos
        if ':' in Results_str:
            parts = Results_str.split(':')
            if len(parts) == 2:  # MM:SS.ms
                minutes, seconds = parts
                return float(minutes) * 60 + float(seconds)
            elif len(parts) == 3:  # HH:MM:SS.ms
                hours, minutes, seconds = parts
                return float(hours) * 3600 + float(minutes) * 60 + float(seconds)
        else:
            # Apenas segundos
            return float(Results_str)
    except:
        return np.nan

# Aplicação da conversão de tempo
print("Convertendo tempos para segundos...")

# Verifica se a coluna 'Results' existe nos dados
if 'Results' in olympic_data.columns:
    olympic_data['Results_Seconds'] = olympic_data['Results'].apply(convert_Results_to_seconds)
elif 'Results_Seconds' not in olympic_data.columns:
    print("Aviso: Coluna 'Results' não encontrada nos dados olímpicos.")

if 'Results' in championship_data.columns:
    championship_data['Results_Seconds'] = championship_data['Results'].apply(convert_Results_to_seconds)
elif 'Results_Seconds' not in championship_data.columns:
    print("Aviso: Coluna 'Results' não encontrada nos dados dos campeonatos.")

# %%
# Criação de chaves de evento para correspondência entre competições
print("Criando chaves de evento...")

# Função para criar chave de evento
def create_event_key(row):
    try:
        # Verifica quais colunas estão disponíveis
        gender = row.get('Gender', '')
        distance = row.get('Distance', row.get('Distance (in meters)', ''))
        stroke = row.get('Stroke', '')
        relay = 'Relay' if row.get('Relay?', False) else 'Individual'

        # Cria a chave
        return f"{gender}_{distance}_{stroke}_{relay}".replace(' ', '_')
    except:
        return "Unknown_Event"

# Aplicação da criação de chaves
olympic_data['EventKey'] = olympic_data.apply(create_event_key, axis=1)
championship_data['EventKey'] = championship_data.apply(create_event_key, axis=1)

# %%
# Mapeamento entre anos olímpicos e campeonatos mundiais anteriores
olympic_to_world_mapping = {
    1976: 1975,
    1980: 1978,
    1984: 1982,
    1988: 1986,
    1992: 1991,
    1996: 1994,
    2000: 1998,
    2004: 2003,
    2008: 2007,
    2012: 2011,
    2016: 2015,
    2020: 2019,
    2024: 2023
}

# Exibição do mapeamento
print("Mapeamento entre anos olímpicos e campeonatos mundiais anteriores:")
mapping_df = pd.DataFrame(list(olympic_to_world_mapping.items()),
                         columns=['Olimpíada', 'Campeonato Mundial Anterior'])
display(mapping_df)

# %%
# Filtragem e preparação dos dados
print("Preparando os dados para modelagem...")

# Filtragem para incluir apenas os anos mapeados
valid_olympic_years = list(olympic_to_world_mapping.keys())
olympic_filtered = olympic_data[olympic_data['Year'].isin(valid_olympic_years)].copy()

# Criação do conjunto de dados integrado
processed_data = []

# Para cada ano olímpico e evento
for olympic_year in valid_olympic_years:
    world_year = olympic_to_world_mapping[olympic_year]

    # Filtra dados olímpicos para o ano atual
    olympic_year_data = olympic_filtered[olympic_filtered['Year'] == olympic_year]

    # Para cada evento nas olimpíadas deste ano
    for event_key in olympic_year_data['EventKey'].unique():
        # Filtra dados olímpicos para este evento
        olympic_event_data = olympic_year_data[olympic_year_data['EventKey'] == event_key]

        # Filtra dados do campeonato mundial para o ano correspondente e mesmo evento
        world_event_data = championship_data[
            (championship_data['Year'] == world_year) &
            (championship_data['EventKey'] == event_key)
        ]

        # Se temos dados para ambos (olimpíadas e mundial)
        if not olympic_event_data.empty and not world_event_data.empty:
            # Obtém os tempos dos três primeiros colocados em cada competição
            try:
                # Dados olímpicos
                olympic_top3 = olympic_event_data.sort_values('Rank').head(3)
                if len(olympic_top3) < 3:
                    continue  # Pula se não tivermos os 3 primeiros

                olympic_rank1 = olympic_top3.iloc[0]
                olympic_rank2 = olympic_top3.iloc[1]
                olympic_rank3 = olympic_top3.iloc[2]

                # Dados do mundial
                world_top3 = world_event_data.sort_values('Rank').head(3)
                if len(world_top3) < 3:
                    continue  # Pula se não tivermos os 3 primeiros

                world_rank1 = world_top3.iloc[0]
                world_rank2 = world_top3.iloc[1]
                world_rank3 = world_top3.iloc[2]

                # Cria entrada para o conjunto de dados processado
                entry = {
                    'OlympicYear': olympic_year,
                    'WorldYear': world_year,
                    'EventKey': event_key,
                    'Gender': olympic_rank1.get('Gender', ''),
                    'Distance (in meters)': olympic_rank1.get('Distance', olympic_rank1.get('Distance (in meters)', '')),
                    'Stroke': olympic_rank1.get('Stroke', ''),
                    'Relay?': olympic_rank1.get('Relay?', False),

                    # Tempos do Mundial
                    'Prev_Champs_Rank1_Results': world_rank1.get('Results_Seconds', np.nan),
                    'Prev_Champs_Rank2_Results': world_rank2.get('Results_Seconds', np.nan),
                    'Prev_Champs_Rank3_Results': world_rank3.get('Results_Seconds', np.nan),

                    # Tempos Olímpicos (alvos)
                    'Olympic_Rank1_Results': olympic_rank1.get('Results_Seconds', np.nan),
                    'Olympic_Rank2_Results': olympic_rank2.get('Results_Seconds', np.nan),
                    'Olympic_Rank3_Results': olympic_rank3.get('Results_Seconds', np.nan),
                }

                processed_data.append(entry)
            except Exception as e:
                print(f"Erro ao processar evento {event_key} para o ano {olympic_year}: {e}")

# Criação do DataFrame processado
processed_df = pd.DataFrame(processed_data)

# Remoção de linhas com valores ausentes
processed_df.dropna(inplace=True)

# Exibição do DataFrame processado
print(f"Conjunto de dados processado criado com {len(processed_df)} entradas.")
display(processed_df.head())

# Salvar o DataFrame processado para uso posterior
processed_df.to_csv('processed_swimming_data.csv', index=False)
print("Dados processados salvos em 'processed_swimming_data.csv'")

# %% [markdown]
# ## 4. Análise Exploratória dos Dados

# %%
# Estatísticas descritivas
print("Estatísticas descritivas dos dados processados:")
display(processed_df.describe())

# %%
# Visualização da correlação entre tempos do mundial e tempos olímpicos
plt.figure(figsize=(18, 6))

# Medalha de ouro
plt.subplot(1, 3, 1)
plt.scatter(processed_df['Prev_Champs_Rank1_Results'], processed_df['Olympic_Rank1_Results'], alpha=0.7)
plt.plot([processed_df['Prev_Champs_Rank1_Results'].min(), processed_df['Prev_Champs_Rank1_Results'].max()],
         [processed_df['Prev_Champs_Rank1_Results'].min(), processed_df['Prev_Champs_Rank1_Results'].max()],
         'r--', label='Linha de Igualdade')
plt.xlabel('Tempo do 1º Colocado no Mundial (segundos)')
plt.ylabel('Tempo do 1º Colocado na Olimpíada (segundos)')
plt.title('Correlação: 1º Colocado Mundial vs. Olimpíada')
plt.legend()

# Medalha de prata
plt.subplot(1, 3, 2)
plt.scatter(processed_df['Prev_Champs_Rank2_Results'], processed_df['Olympic_Rank2_Results'], alpha=0.7)
plt.plot([processed_df['Prev_Champs_Rank2_Results'].min(), processed_df['Prev_Champs_Rank2_Results'].max()],
         [processed_df['Prev_Champs_Rank2_Results'].min(), processed_df['Prev_Champs_Rank2_Results'].max()],
         'r--', label='Linha de Igualdade')
plt.xlabel('Tempo do 2º Colocado no Mundial (segundos)')
plt.ylabel('Tempo do 2º Colocado na Olimpíada (segundos)')
plt.title('Correlação: 2º Colocado Mundial vs. Olimpíada')
plt.legend()

# Medalha de bronze
plt.subplot(1, 3, 3)
plt.scatter(processed_df['Prev_Champs_Rank3_Results'], processed_df['Olympic_Rank3_Results'], alpha=0.7)
plt.plot([processed_df['Prev_Champs_Rank3_Results'].min(), processed_df['Prev_Champs_Rank3_Results'].max()],
         [processed_df['Prev_Champs_Rank3_Results'].min(), processed_df['Prev_Champs_Rank3_Results'].max()],
         'r--', label='Linha de Igualdade')
plt.xlabel('Tempo do 3º Colocado no Mundial (segundos)')
plt.ylabel('Tempo do 3º Colocado na Olimpíada (segundos)')
plt.title('Correlação: 3º Colocado Mundial vs. Olimpíada')
plt.legend()

plt.tight_layout()
plt.savefig('correlacao_tempos.png')
plt.show()

# %%
# Distribuição de provas por estilo
plt.figure(figsize=(12, 6))
stroke_counts = processed_df['Stroke'].value_counts()
sns.barplot(x=stroke_counts.index, y=stroke_counts.values)
plt.title('Distribuição de Provas por Estilo')
plt.xlabel('Estilo')
plt.ylabel('Número de Provas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('distribuicao_estilos.png')
plt.show()

# %%
# Evolução dos tempos médios ao longo dos anos
plt.figure(figsize=(14, 8))

# Agrupamento por ano olímpico e cálculo da média dos tempos
Results_evolution = processed_df.groupby('OlympicYear')[
    ['Olympic_Rank1_Results', 'Olympic_Rank2_Results', 'Olympic_Rank3_Results']
].mean()

# Plotagem
plt.plot(Results_evolution.index, Results_evolution['Olympic_Rank1_Results'], 'o-', label='1º Colocado')
plt.plot(Results_evolution.index, Results_evolution['Olympic_Rank2_Results'], 's-', label='2º Colocado')
plt.plot(Results_evolution.index, Results_evolution['Olympic_Rank3_Results'], '^-', label='3º Colocado')

plt.title('Evolução dos Tempos Médios nas Olimpíadas')
plt.xlabel('Ano Olímpico')
plt.ylabel('Tempo Médio (segundos)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('evolucao_tempos.png')
plt.show()

# %% [markdown]
# ## 5. Modelagem Preditiva

# %%
# Definição das variáveis para modelagem
TARGET_COLS = ["Olympic_Rank1_Results", "Olympic_Rank2_Results", "Olympic_Rank3_Results"]
FEATURE_COLS = ["Prev_Champs_Rank1_Results", "Prev_Champs_Rank2_Results", "Prev_Champs_Rank3_Results",
                "OlympicYear", "Distance (in meters)", "Stroke", "Gender", "Relay?"]
CATEGORICAL_FEATURES = ["Distance (in meters)", "Stroke", "Gender", "Relay?"]
NUMERICAL_FEATURES = ["Prev_Champs_Rank1_Results", "Prev_Champs_Rank2_Results", "Prev_Champs_Rank3_Results", "OlympicYear"]
TEST_YEAR_THRESHOLD = 2000  # Usar todas as Olimpíadas a partir de 2000 como conjunto de teste

# Criação do diretório para resultados
os.makedirs('ml_Results', exist_ok=True)
Results_DIR = 'ml_Results'

# %%
# Separação dos dados (abordagem temporal)
print(f"Separando dados: anos olímpicos anteriores a {TEST_YEAR_THRESHOLD} para treino, e a partir de {TEST_YEAR_THRESHOLD} (inclusive) para teste...")
df_train = processed_df[processed_df["OlympicYear"] < TEST_YEAR_THRESHOLD].copy()
df_test = processed_df[processed_df["OlympicYear"] >= TEST_YEAR_THRESHOLD].copy()

# Verificação da separação
if df_train.empty or df_test.empty:
    print(f"Aviso: Conjunto de treino ou teste vazio após separação pelo ano limite {TEST_YEAR_THRESHOLD}.")
    print(f"Tamanho do treino: {df_train.shape}, Tamanho do teste: {df_test.shape}")
    # Fallback para divisão aleatória se a separação temporal falhar
    print("Utilizando divisão aleatória como alternativa...")
    if len(processed_df) < 2:
        print("Erro: Não há dados suficientes nem para uma divisão aleatória.")
        raise ValueError("Conjunto de dados insuficiente para modelagem")
    df_train, df_test = train_test_split(processed_df, test_size=0.2, random_state=42)
    if df_train.empty or df_test.empty:
        print("Erro: Conjunto de treino ou teste vazio mesmo após divisão aleatória.")
        raise ValueError("Falha na separação dos dados")

X_train = df_train[FEATURE_COLS]
X_test = df_test[FEATURE_COLS]

print(f"Tamanho do conjunto de treino: {X_train.shape[0]}, Tamanho do conjunto de teste: {X_test.shape[0]}")

# %%
# Configuração do pré-processamento
print("Configurando pipeline de pré-processamento...")
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), NUMERICAL_FEATURES),
        ("cat", OneHotEncoder(handle_unknown='ignore'), CATEGORICAL_FEATURES)
    ],
    remainder='passthrough'  # Mantém outras colunas, se houver
)

# Definição dos modelos
models = {
    "LinearRegression": LinearRegression(),
    "DecisionTree": DecisionTreeRegressor(random_state=42),
    "RandomForest": RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1),
    "XGBoost": xgb.XGBRegressor(objective='reg:squarederror', random_state=42, n_jobs=-1),
    "MLP": MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42, early_stopping=True)
}

# %%
# Loop de treinamento e avaliação
Results = {}
predictions_all = {}  # Armazenar previsões para visualização

for target_col in TARGET_COLS:
    print(f"\n--- Treinando modelos para o alvo: {target_col} ---")
    y_train = df_train[target_col]
    y_test = df_test[target_col]
    Results[target_col] = {}
    predictions_all[target_col] = df_test[["OlympicYear", "EventKey"]].copy()
    predictions_all[target_col]["Actual_Results"] = y_test

    for model_name, model in models.items():
        print(f"Treinando {model_name}...")

        # Criação do pipeline
        pipeline = Pipeline(steps=[("preprocessor", preprocessor), ("regressor", model)])

        try:
            # Treinamento do modelo
            pipeline.fit(X_train, y_train)

            # Previsão
            y_pred = pipeline.predict(X_test)
            predictions_all[target_col][f"{model_name}_Pred_Results"] = y_pred

            # Avaliação
            mae = mean_absolute_error(y_test, y_pred)
            mse = mean_squared_error(y_test, y_pred)
            rmse = np.sqrt(mse)
            r2 = r2_score(y_test, y_pred)

            Results[target_col][model_name] = {
                "MAE": mae,
                "MSE": mse,
                "RMSE": rmse,
                "R2": r2
            }
            print(f"{model_name} - MAE: {mae:.4f}, RMSE: {rmse:.4f}, R2: {r2:.4f}")

        except Exception as e:
            print(f"Erro ao treinar/avaliar {model_name} para {target_col}: {e}")
            Results[target_col][model_name] = {"Error": str(e)}

# %%
# Salvamento dos resultados
print("\nSalvando resultados da avaliação...")
Results_df = pd.DataFrame.from_dict({(i,j): Results[i][j]
                                   for i in Results.keys()
                                   for j in Results[i].keys()},
                                  orient='index')
Results_df.index = pd.MultiIndex.from_tuples(Results_df.index, names=["TargetRank", "Model"])
Results_df.to_csv(os.path.join(Results_DIR, "model_evaluation_metrics.csv"))
print(f"Métricas de avaliação salvas em {os.path.join(Results_DIR, 'model_evaluation_metrics.csv')}")

# Salvamento das previsões
print("Salvando previsões...")
for target_col, pred_df in predictions_all.items():
    pred_df.to_csv(os.path.join(Results_DIR, f"predictions_{target_col}.csv"), index=False)
    print(f"Previsões para {target_col} salvas em {os.path.join(Results_DIR, f'predictions_{target_col}.csv')}")

# %% [markdown]
# ## 6. Visualização e Comparação dos Resultados

# %%
# Função para criar gráficos de comparação de métricas
def plot_metric_comparison(Results_df, metric):
    plt.figure(figsize=(14, 8))

    # Filtragem para a métrica específica
    metric_df = Results_df.reset_index()
    metric_df = metric_df[metric_df.columns[metric_df.columns.str.contains(metric)]]

    # Preparação para o gráfico
    plot_data = pd.DataFrame()
    plot_data['Model'] = Results_df.index.get_level_values('Model')
    plot_data['TargetRank'] = Results_df.index.get_level_values('TargetRank')
    plot_data[metric] = metric_df.values

    # Criação do gráfico
    sns.barplot(data=plot_data, x="Model", y=metric, hue="TargetRank", palette="viridis")
    plt.title(f"Comparação de {metric} por Modelo e Posição no Pódio Olímpico")
    plt.ylabel(metric)
    plt.xlabel("Modelo")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(os.path.join(Results_DIR, f"comparison_{metric}.png"))
    plt.show()

# Criação dos gráficos de comparação
for metric in ["MAE", "RMSE", "R2"]:
    plot_metric_comparison(Results_df, metric)

# %%
# Função para criar gráficos de dispersão (valores reais vs. previstos)
def plot_actual_vs_predicted(predictions_df, model_name, target_col):
    plt.figure(figsize=(10, 8))

    # Colunas para o gráfico
    actual_col = "Actual_Results"
    pred_col = f"{model_name}_Pred_Results"

    # Verificação das colunas
    if pred_col not in predictions_df.columns:
        print(f"Aviso: Coluna {pred_col} não encontrada em {predictions_df.columns}")
        return

    # Criação do gráfico
    sns.scatterplot(data=predictions_df, x=actual_col, y=pred_col)

    # Linha de igualdade (y=x)
    min_val = min(predictions_df[actual_col].min(), predictions_df[pred_col].min())
    max_val = max(predictions_df[actual_col].max(), predictions_df[pred_col].max())
    plt.plot([min_val, max_val], [min_val, max_val], color='red', linestyle='--', label='Linha Ideal (y=x)')

    plt.title(f"Valores Reais vs. Previstos: {model_name} para {target_col}")
    plt.xlabel("Tempo Real (segundos)")
    plt.ylabel("Tempo Previsto (segundos)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(Results_DIR, f"actual_vs_pred_{model_name}_{target_col}.png"))
    plt.show()

# Criação dos gráficos de dispersão para o melhor modelo (LinearRegression)
for target_col, pred_df in predictions_all.items():
    plot_actual_vs_predicted(pred_df, "LinearRegression", target_col)

# %% [markdown]
# ## 7. Resumo e Conclusões

# %%
# Criação de um resumo dos resultados
print("Resumo dos Resultados:")
print("-" * 50)

# Melhor modelo por métrica
best_mae_model = Results_df['MAE'].idxmin()
best_rmse_model = Results_df['RMSE'].idxmin()
best_r2_model = Results_df['R2'].idxmax()

print(f"Melhor modelo por MAE: {best_mae_model[1]} para {best_mae_model[0]} (MAE = {Results_df.loc[best_mae_model, 'MAE']:.4f})")
print(f"Melhor modelo por RMSE: {best_rmse_model[1]} para {best_rmse_model[0]} (RMSE = {Results_df.loc[best_rmse_model, 'RMSE']:.4f})")
print(f"Melhor modelo por R²: {best_r2_model[1]} para {best_r2_model[0]} (R² = {Results_df.loc[best_r2_model, 'R2']:.4f})")

print("\nDesempenho médio dos modelos:")
model_avg = Results_df.groupby('Model').mean()
display(model_avg)

# Salvamento do resumo
with open(os.path.join(Results_DIR, "model_comparison_summary.txt"), "w") as f:
    f.write("Resumo da Comparação dos Modelos\n")
    f.write("=" * 50 + "\n\n")

    f.write("Melhores Modelos por Métrica:\n")
    f.write(f"- Melhor modelo por MAE: {best_mae_model[1]} para {best_mae_model[0]} (MAE = {Results_df.loc[best_mae_model, 'MAE']:.4f})\n")
    f.write(f"- Melhor modelo por RMSE: {best_rmse_model[1]} para {best_rmse_model[0]} (RMSE = {Results_df.loc[best_rmse_model, 'RMSE']:.4f})\n")
    f.write(f"- Melhor modelo por R²: {best_r2_model[1]} para {best_r2_model[0]} (R² = {Results_df.loc[best_r2_model, 'R2']:.4f})\n\n")

    f.write("Desempenho Médio dos Modelos:\n")
    f.write(model_avg.to_string() + "\n\n")

    f.write("Conclusões:\n")
    f.write("1. A Regressão Linear apresentou desempenho superior em todas as métricas avaliadas.\n")
    f.write("2. Existe uma relação predominantemente linear entre os tempos dos mundiais e os subsequentes tempos olímpicos.\n")
    f.write("3. Os modelos conseguem prever com maior precisão os tempos do primeiro colocado (medalha de ouro).\n")
    f.write("4. A abordagem de separação temporal dos dados garantiu uma avaliação realista da capacidade preditiva dos modelos.\n")
    f.write("5. Os resultados sugerem que é possível fazer projeções confiáveis dos tempos olímpicos com base nos resultados do campeonato mundial anterior.\n")

print(f"\nResumo salvo em {os.path.join(Results_DIR, 'model_comparison_summary.txt')}")

# %% [markdown]
# ## 8. Download dos Resultados

# %%
# Preparação dos arquivos para download
print("Preparando arquivos para download...")

# Lista de arquivos para download
files_to_download = [
    os.path.join(Results_DIR, "model_evaluation_metrics.csv"),
    os.path.join(Results_DIR, "model_comparison_summary.txt"),
    os.path.join(Results_DIR, "comparison_R2.png"),
    os.path.join(Results_DIR, "comparison_MAE.png"),
    os.path.join(Results_DIR, "comparison_RMSE.png"),
    os.path.join(Results_DIR, "actual_vs_pred_LinearRegression_Olympic_Rank1_Results.png")
]

# Download dos arquivos
for file_path in files_to_download:
    try:
        files.download(file_path)
        print(f"Download disponível para: {file_path}")
    except Exception as e:
        print(f"Erro ao preparar download para {file_path}: {e}")

print("\nTodos os arquivos estão disponíveis para download. Clique nos links acima para baixá-los.")

# %% [markdown]
# ## Conclusão
#
# Este notebook implementou uma análise preditiva completa dos resultados olímpicos de natação com base nos resultados dos campeonatos mundiais anteriores. Os modelos desenvolvidos demonstraram alta capacidade preditiva, com a Regressão Linear apresentando desempenho superior em todas as métricas avaliadas.
#
# Os resultados sugerem uma relação predominantemente linear entre os tempos dos mundiais e os subsequentes tempos olímpicos, o que tem implicações práticas significativas para treinadores, atletas e comitês olímpicos na preparação para competições futuras.
#
# A metodologia empregada, especialmente a abordagem de separação temporal dos dados, garantiu uma avaliação realista da capacidade preditiva dos modelos, simulando um cenário de aplicação prática onde se deseja prever resultados de competições futuras com base em dados históricos.
```
